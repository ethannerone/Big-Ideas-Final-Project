In this data set I will get the lowest crime rate in pittsburgh. The reason that I am doing this is that it is better to be in a low crime area during an purgatory event. I will get the lowest crime by getting population data on each neighborhood and comparing that to the amount of crimes that had occured in each neighborhood.

```python
#Here I simply print out the instances of crimes that occur in each neighborhood in pittsburgh. I do this by reading through all
#the crime data and incrementing to each neighborhood at each instance of a crime.
dict = {}
import pandas as pd
crime = pd.read_csv("Crimedata.csv")
# I use a for loop to go through every instance of a crime and adds it to a dictionary if it is unique, or increments if it already
# exists.
for i in range(2936):
    if crime.iloc[i,6] in dict:
        dict[crime.iloc[i,6]] = dict[crime.iloc[i,6]] + 1
    else:
        dict.update({crime.iloc[i,6] : 1})
print(dict)
```

```python
dict = {}
import pandas as pd
crime = pd.read_csv("Crimedata.csv")
# I use the same for loop I used to get the amount of crime in each neighborhood
for i in range(2936):
    if crime.iloc[i,6] in dict:
        dict[crime.iloc[i,6]] = dict[crime.iloc[i,6]] + 1
    else:
        dict.update({crime.iloc[i,6] : 1})
crimevalues = {}
pop = pd.read_csv("Neighborhood.csv")
# I use a different for loop that iterates through each neighborhood, and it's repsective popuation. After I do this I 
# get the ratio of crime per neighborhood by dividing the population of each neighborhood by the number of crimes commited there.
for i in range(89):
    if pop.iloc[i,0] in dict:
        crimevalues.update({pop.iloc[i,0]: (dict[pop.iloc[i,0]])/(pop.iloc[i,1])})
print(crimevalues)
```

```python
import pandas as pd
import matplotlib.pylab as plt
dict = {}
crime = pd.read_csv("Crimedata.csv")
#I use all the same code that I did in the previous problem with some minor modifications
for i in range(2936):
    if crime.iloc[i,6] in dict:
        dict[crime.iloc[i,6]] = dict[crime.iloc[i,6]] + 1
    else:
        dict.update({crime.iloc[i,6] : 1})
crimevalues = {}
pop = pd.read_csv("Neighborhood.csv")
for i in range(89):
    if pop.iloc[i,0] in dict:
        # Now I only add very low crime areas to my crimevalues dictionary
        if (dict[pop.iloc[i,0]])/(pop.iloc[i,1]) < 0.00265:
            crimevalues.update({pop.iloc[i,0]: (dict[pop.iloc[i,0]])/(pop.iloc[i,1])})
# I initialize the plot by setting it to the items in my crimevalues dictionary
myList = crimevalues.items()
myList = sorted(myList) 
x, y = zip(*myList) 
# I plot my graph.
plt.plot(x, y)
plt.show()
```

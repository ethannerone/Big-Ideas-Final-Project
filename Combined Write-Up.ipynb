# Project Topic: Which Pittsburgh neighborhood is the best to survive the Purge?
(From the movie franchice The Purge)
For our Final Project, we decided to research which Pittsburgh neighborhood is the best to survive the Purge, the fictitious event from the cinematic franchise "The Purge." Essentially, it is a 12 hour event in the United States from 7:00pm to 7:00am where all crime is legal and emergency services (such as ambulances, police and the fire department). Additionally:

- Sirens blare throughout the land to signal the start and end of The Purge.
- All police, fire, and emergency medical services are suspended for the duration of the Purge.
- The president, first family, and government officials of rank 10, are granted immunity from the Purge, meaning that they must not be harmed. (This was revoked in Election Year to permit the murder of a political opponent).
- Only Class 14 weaponry is permitted, while Class 5 weaponry (such as explosives) is prohibited.
- Violations of any Purge rules will result in public execution. And since Pittsburgh is in the United States, our inquiry is which Pittsburghs - neighborhood is the best to survive the Purge in?

The number one threat during the Purge are malicious individuals or groups who want to ransack and raid various places, to either steal, sabotage or worse, kill. The best way to avoid these is to hunker in a safe place, in a relatively safe area with more valuable buildings that malicious individuals would rather target over your's. The neighborhoods that provides the best of these will be calculated with the following metrics:

|Metric|Reasoning|
|---|---|
|Crime rates per neighborhood|The crime rate of a neighborhood outside of the Purge is very telling of the types of individuals living in the population; the more crime a neighborhood has, the more criminals or wannabes a neighborhood has, whom might want to perform malicious acts during the Purge. The less crime a neighborhood has, the less malicious people will roam during the Purge, and the less chance one such individual will encounter you.|
|Grocery store to population ratio|Groceries store serve as two purposes during the Purge: emergency supplies and distractions to draw away malicious people away from your hiding spot. The more grocery stores a neighborhood has, the higher density and less potential walking distance from your hiding spot if you require any emergency supplies. And alternatively the more grocery stores means that malicious people have more valuable places as options to raid when compared to your hiding place. Additionally, we look at the total population of each neighborhood, to give us an idea of how many people could attack every grocery store as a worst-case-scenario.|
|Amount of city facilities|These city-owned facilities include amongst them many useful and safe hiding spots perfect for surviving the Purge. Examples of such include police stations with armories, sturdy fire departments and disaster-proof shelters. But city facilities can also include not-so-safe facilities such as concession stands, public swimming pools and baseball fields, all of which provide no safety from malicious people. Therefore, we will instead evaluate the number of useful/safe city facilities in a city. The more a neighborhood has, the more options to hide and the closer an alternative will be incase your primary hiding spot is compromised.|

## Reasoning
